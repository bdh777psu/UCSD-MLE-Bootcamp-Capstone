'Mental health in the tech industry' is a complex and increasingly important issue as it encompasses many challenges with potentially severe impacts to one's well being: Fast-paced environments, high pressure, poor work-life balance, imposter syndrome, and lack of diversity can all play a significant role towards employee turnover, developing mental health issues, and even burnout.





Before starting to write any code and building the first model prototype for the Capstone project, what follows is an attempt to locate published research around or on the topic of Mental Health in the Tech Industry: The goal is to understand what other researchers/engineers have built to solve a similar problem.

As a non-profit, 501(c)(3) corporation, 'Open Sourcing Mental Health' (OSMH) is dedicated to raising awareness, educating, and providing resources to support mental wellness in the tech and open source communities, OSMH's mission is to change how we talk about mental health in the tech community, by raising awareness, and providing education and resources on this topic: https://osmhhelp.org/index.html
