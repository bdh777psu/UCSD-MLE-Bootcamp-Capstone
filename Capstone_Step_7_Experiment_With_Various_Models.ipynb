The dataset contains informative data from countries across the globe about the prevalence of mental health disorders including schizophrenia, bipolar disorder, eating disorders, anxiety disorders, drug use disorders, depression and alcohol use disorders. By providing this data in an easy to visualise format you can gain an insight into how these issues are impacting lives; allowing for a deeper understanding of these conditions and the implications.p

For this capstone step I'll be using the [Global Trends in Mental Health Disorder dataset](https://www.kaggle.com/datasets/thedevastator/uncover-global-trends-in-mental-health-disorder). First, let's import all the libraries we'll be using.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score
```

```python
# Load the dataset into a pandas DataFrame
data = pd.read_csv('/content/drive/MyDrive/UCSD-ML/Capstone/raw_data/Mental_health_Depression_disorder_Data.csv', low_memory=False)
```

```python
# Display the first few rows of the dataset
data.head()
```

```python
# Drop rows with missing values.
data.dropna(inplace=True)
```

```python
feature_cols = ['Schizophrenia (%)',	'Bipolar disorder (%)',	'Eating disorders (%)',	'Anxiety disorders (%)', 'Drug use disorders (%)',	'Depression (%)',	'Alcohol use disorders (%)']
```

```python
# Split the dataset into training and testing sets
X = data[feature_cols]
y = data['Entity']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
# Create a Decision Tree model
from sklearn.tree import DecisionTreeClassifier

clf_tree = DecisionTreeClassifier()
clf_tree.fit(X_train, y_train)
```

```python
# Create a Random Forest model
from sklearn.ensemble import RandomForestClassifier

clf_forest = RandomForestClassifier()
clf_forest.fit(X_train, y_train)
```

```python
# Extract feature importances
feature_importances = pd.DataFrame({'feature_name': clf_forest.feature_names_in_, 'feature_importance': clf_forest.feature_importances_}).sort_values(by='feature_importance', ascending=False)
```

```python
# Print feature importance
feature_importances
```

```python
# Build and train an AdaBoostClassifier on your training data.
from sklearn.ensemble import AdaBoostClassifier
clf_boost = AdaBoostClassifier(estimator=DecisionTreeClassifier(max_depth=1))
clf_boost.fit(X_train, y_train)
```

```python
# Evaluate models
from sklearn.model_selection import cross_validate
from sklearn.metrics import accuracy_score

clf_tree_accuracy = cross_validate(clf_tree, X, y, scoring='accuracy')
print(clf_tree_accuracy['test_score'].mean())

clf_forest_accuracy = cross_validate(clf_forest, X, y, scoring='accuracy')
print(clf_forest_accuracy['test_score'].mean())

clf_boost_accuracy = cross_validate(clf_boost, X, y, scoring='accuracy')
print(clf_boost_accuracy['test_score'].mean())
```
